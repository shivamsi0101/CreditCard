# Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# it is use for read the data set
creditcard = pd.read_csv("creditcard.csv")
```

# Fraudlent Transaction

```python
legit = creditcard[creditcard.Class == 0]
fraud = creditcard[creditcard.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#to show the amount of transaction
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#group of all classes in dataset
creditcard.groupby('Class').mean()
```

# Under Sampling

```python
legit_sample = legit.sample(n=30)
```

```python
#it show the new dataset after the modify data
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
#it shown the 5 top data of dataset
new_dataset.head()
```

```python
#it shown the 5 last data of dataset
new_dataset.tail()
```

```python
#new dataset
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
# value of new dataset colomn and axis 1
print(X)
```

```python
# value of new dataset class
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
